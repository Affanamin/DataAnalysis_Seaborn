### Why we are using Seaborn? 

Seaborn is a next level of visualization library, certainly built on matplotlib, So it gives all the functionalities which matlpotlib with extra functionalities as well.

Seaborn is complimentary to Matplotlib and it specifically targets statistical data visualization. 


Michael Waskom once said: “If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.”

Seaborn works with different parameters amazingly well, which undoubtedly speaks to those users who  don’t use the default looks of the Matplotlib plots.

Lets Compare Both of them and see the figure's aesthetics

```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Initialize Figure and Axes object
fig, ax = plt.subplots()

# Load in data
tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

# Create violinplot
ax.violinplot(tips["total_bill"], vert=False)

# Show the plot
plt.show()
```

```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
tips = sns.load_dataset("tips")

# Create violinplot
sns.violinplot(x = "total_bill", data=tips)

# Show the plot

plt.show()
```

We can clearly see the difference of both of the pictures with default parameters, Seaborn gives more appealing graphs or figures.

### Limitations of Matplotlib:

1) Deafult parameters doesnot work ver effective in matplot, as you can see in above example.

2) Matplotlib does not works very well with pandas data frames, while seaborn works really well with pandas dataframes.

3) Drawing attractive figures is important, Matplotlib doesnot have very attractive figures to show while seaborn gives a great control on figure aesthetics, Lets see this in below example.

Reference: https://seaborn.pydata.org/tutorial/aesthetics.html

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
        
        
```

```python
#Matplotlib

sinplot()
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sinplot()
```

```python
sns.set_context("poster") #you can select themes for visuals
sinplot()
```
