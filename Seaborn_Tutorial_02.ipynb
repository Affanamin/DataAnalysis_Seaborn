```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_train = pd.read_csv("train.csv")
```

```python
data_train.head()
```

```python
#lets learn some concepts about uni/bi/multi variate data analysis.
```

### Distribution Plots:

```python
#For BiVariate Analysis (Means analysis of 2 indep Features), Use JointPlot

sns.jointplot(x="Age",y="Fare",data = data_train,kind='hex')
```

```python
sns.jointplot(x="Age",y="Fare",data = data_train,kind='reg')
```

```python
## #For MultiVariate Analysis (Means analysis of more than 2 indep Features), Use Pairplot

sns.pairplot(data_train)
```

```python
#permutation and combination of all int /floating features, 
```

```python
## #For Univariate Analysis (Means analysis of 1 indep Features), Use distplot
sns.distplot(data_train['Age'])
```

```python
#It gives the histogram and show distribution of data.
```

```python
sns.distplot(data_train['Age'],kde = False, bins = 10) 

#KDE means In statistics, kernel density estimation (KDE) is a 
# non-parametric way to estimate the probability density function of a random variable. 
```
